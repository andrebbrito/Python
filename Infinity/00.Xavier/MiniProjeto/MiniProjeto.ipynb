# Mini-Projeto
Vamos praticar o que aprendemos até aqui criando um pequeno projeto **CRUD** (Create, Read, Update and Delete) para controlar um **cadastro de clientes**.\
Por enquanto, nosso projeto terá as seguintes limitações:
* Interface totalmente em texto
* Armazenamento volátil (sem persistência)

Em aulas futuras aprenderemos como aprimorar a solução utilizando recursos mais poderosos.


## **Dados do Cadastro de Clientes**


* Nome do cliente (**nome**)
* E-mail do cliente (**email**)
* Tipo do cliente (**tipo**): atacadista ou varejista
* Percentual de desconto que o cliente recebe na loja (**desconto**)


## **Estrutura do aplicativo**


* Menu principal
  - Inclusão de um novo cliente
  - Ateração de dados de um cliente cadastrado
  - Exclusão de um cliente cadastrado
  - Consultas
    - Lista de nomes e e-mails de clientes (todos ou de um único tipo)
    - Dados de um único cliente
    - Clientes acima de um certo percentual

## **Dica**

Que tal criar um dicionário onde cada par chave/valor for um cliente?
- A chave poderia ser uma identificação do cliente (um id qualquer ou o próprio nome)
- O valor de cada chave seria um dicionário contendo chaves para cada informação de um único cliente

Exemplo:

```python
animais = {
    'Rex':{'tipo':'cão','sexo':'m','cor':'caramelo'},
    'Marie':{'tipo':'gato','sexo':'f','cor':'branco'},
    'Paco':{'tipo':'papagaio','sexo':'m','cor':'verde'}
}

print(animais)
print(animais['Rex'])
print(animais['Rex']['cor'])
```
